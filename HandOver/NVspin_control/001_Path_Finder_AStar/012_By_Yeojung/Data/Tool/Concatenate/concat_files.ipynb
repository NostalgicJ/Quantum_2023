```python
import pandas as pd

df1 = pd.read_csv("ByAstar_dt_2.6_part_20231005_113421.csv")
df2 = pd.read_csv("ByAstar_dt_2.6_part_20231005_113433.csv")
df3 = pd.read_csv("ByAstar_dt_2.6_part_20231005_113442.csv")
df4 = pd.read_csv("ByAstar_dt_2.6_part_20231006_103959.csv")
df5 = pd.read_csv("ByAstar_dt_2.6_part_20231006_104100.csv")
df6 = pd.read_csv("ByAstar_dt_2.6_part_20231006_104217.csv")
df7 = pd.read_csv("ByAstar_dt_2.6_part_20231006_175233.csv")
df8 = pd.read_csv("ByAstar_dt_2.6_part_20231006_175318.csv")
```

```python
result = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8])

result.info()
```

```python
import matplotlib as plt

%matplotlib inline

result.hist(bins=50, figsize=(20,15))
```

```python
result.plot(kind = "scatter", x="Theta", y="Phi", alpha=0.5)
```

```python
result_minus_1 = result[result['combination'] == '[-1]']

result_minus_1.info()
```

```python
result_minus_1.plot(kind = "scatter", x="Theta", y="Phi", alpha=0.5)
```

```python
result_preprocessed = result[result['combination'] != '[-1]']

result_preprocessed.info()

result_preprocessed.plot(kind = "scatter", x="Theta", y="Phi", alpha=0.5)
```

```python
import os
from datetime import datetime

# Create a directory to store the results
dir = 'concatenated'
if not os.path.exists(dir):
    os.makedirs(dir)

# CSV file name setup
date = datetime.now()
printdate = date.strftime('%Y%m%d_%H%M%S')

filename = "/data_concatenated_" + printdate + '.csv'

result_preprocessed.to_csv(dir + filename, index=False)
```
