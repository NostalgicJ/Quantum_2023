```python
import pandas as pd

# CSV 파일에서 데이터 읽기
df1 = pd.read_csv('Astar_results_dir/ByAstar_dt_2.6_part_20230927_143016.csv')
df2 = pd.read_csv('Astar_results_dir/ByAstar_dt_2.6_part_20230927_143222.csv')
df3 = pd.read_csv('Astar_results_dir/ByAstar_dt_2.6_part_20230927_143347.csv')
df4 = pd.read_csv('Astar_results_dir/ByAstar_dt_2.6_part_20230927_143625.csv')
df5 = pd.read_csv('Astar_results_dir/ByAstar_dt_2.6_part_20230927_143708.csv')

df = pd.concat([df1,df2,df3,df4,df5])

df.info()
# df.head()
```

```python
import matplotlib as plt

%matplotlib inline
df.hist(bins=50, figsize=(20,15))

# 히스토그램(주어진 값 범위 수평축, 샘플 수 수직축)
```

```python
df.plot(kind = "scatter", x="Theta", y="Phi", alpha=0.5)
```

```python
df_min1 = df[df['combination'] == '[-1]']

df_min1.info()

df_min1.plot(kind = "scatter", x="Theta", y="Phi", alpha=0.5)
```

```python
df_non_min1 = df[df['combination'] != '[-1]']

df_non_min1.info()

df_non_min1.plot(kind = "scatter", x="Theta", y="Phi", alpha=0.5)
```
