```python
from itertools import product

original_list = [3, 1, 4, 2, 3]


max_value = 5  # 각 위치에 대해 가능한 최댓값

for combination in product(range(max_value), repeat=len(original_list) - 1):
    current_list = original_list[:-1] + list(combination)
    print(current_list)
```

```python
from itertools import product

my_list = [3, 1]
max_value = 5  # 각 위치에 대해 가능한 최댓값

# 가능한 모든 조합을 생성하고 출력
for combination in product(range(max_value), repeat=len(my_list)):
    my_list[:] = combination
    print(my_list)
```

```python
from itertools import product

my_list = [3, 1, 4, 2, 3]
max_value = 5  # 각 위치에 대해 가능한 최댓값

# 가능한 모든 조합을 생성하고 출력
for combination in reversed(list(product(range(max_value), repeat=len(my_list)))):
    my_list[:] = combination
    print(my_list)
```

```python
from itertools import product

my_list = [3, 1, 4, 2, 3]
max_value = 5  # 각 위치에 대해 가능한 최댓값

# 가능한 모든 조합을 뒤집어서 생성하고 출력
for combination in reversed(list(product(range(max_value), repeat=len(my_list)))):
    my_list[-len(combination):] = combination
    print(my_list)
```
